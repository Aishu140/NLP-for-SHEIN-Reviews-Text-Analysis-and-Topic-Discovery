```python
import pandas as pd
import re
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
from gensim import corpora, models
from nltk.corpus import stopwords
import pyLDAvis
import pyLDAvis.gensim_models
```

```python
import nltk
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

# Data
data = {
    "Review_ID": range(1, 16),
    "Review_Text": [
        "Love the dress! Perfect fit and excellent quality.",
        "The delivery was late, but the product is good.",
        "Sizes run small, had to return one item.",
        "Great selection of clothes at affordable prices.",
        "Customer service was very helpful and polite.",
        "Fabric quality is poor, not worth the price.",
        "Everything arrived on time and nicely packed.",
        "Item looks different from the picture, slightly disappointed.",
        "Amazing shoes! Comfortable and stylish.",
        "Had issues with payment, but resolved quickly by support.",
        "Received the wrong color, had to exchange.",
        "Fast shipping and good packaging, very satisfied.",
        "Stitching came loose after one wash, poor durability.",
        "Love the patterns and colors, perfect for summer.",
        "Sizing chart is misleading, need to check carefully."
    ],
    "Rating": [5, 4, 2, 5, 5, 1, 5, 3, 5, 4, 2, 5, 1, 5, 2],
    "Sentiment": [
        "Positive", "Neutral", "Negative", "Positive", "Positive",
        "Negative", "Positive", "Neutral", "Positive", "Neutral",
        "Negative", "Positive", "Negative", "Positive", "Negative"
    ]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Stopwords
stop_words = set(stopwords.words('english'))

# Preprocessing
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = nltk.word_tokenize(text)
    tokens = [w for w in tokens if w not in stop_words]
    return tokens

processed_docs = [preprocess(doc) for doc in df['Review_Text']]

print("Preprocessed Documents:", processed_docs)

# TF-IDF Analysis
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['Review_Text'])
tfidf_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

for i, doc in enumerate(tfidf_df.values):
    top_indices = doc.argsort()[-10:][::-1]
    top_words = [(vectorizer.get_feature_names_out()[j], doc[j]) for j in top_indices if doc[j] > 0]
    print(f"\nTop words in Document {i+1}: {top_words}")

# Word2Vec Embeddings
model = Word2Vec(sentences=processed_docs, vector_size=50, window=5, min_count=1, workers=4)

# Similar words
print("\nSimilar words to 'dress':", model.wv.most_similar("dress", topn=5))
print("Similar words to 'quality':", model.wv.most_similar("quality", topn=5))

words = list(model.wv.index_to_key)
vectors = [model.wv[w] for w in words]
pca = PCA(n_components=2)
x_vals = pca.fit_transform(vectors)

plt.figure(figsize=(8,6))
plt.scatter(x_vals[:,0], x_vals[:,1])
for i, word in enumerate(words):
    plt.annotate(word, (x_vals[i,0], x_vals[i,1]))
plt.title("Word2Vec Embeddings (PCA)")
plt.show()

# LDA Modeling
dictionary = corpora.Dictionary(processed_docs)
corpus = [dictionary.doc2bow(text) for text in processed_docs]

lda_model = models.LdaModel(corpus, num_topics=3, id2word=dictionary, passes=10)

print("\nLDA Topics:")
for idx, topic in lda_model.print_topics(num_words=5):
    print(f"Topic {idx+1}: {topic}")

# Assign topic to each document
for i, row in enumerate(lda_model[corpus]):
    row = sorted(row, key=lambda x: x[1], reverse=True)
    print(f"Document {i+1} -> Topic {row[0][0]+1}")

# pyLDAvis visualization
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)
pyLDAvis.save_html(vis, 'lda_topics.html')

print("LDA visualization is saved as lda_topics.html, this file should be opened in the browser.")
```
